To identify fields or topics users are interested in, I extract meaningful keywords from the "Name" column and then check if these keywords are present in predefined lists of keywords that correspond to different fields or topics.

```python
# Firstly, In order to predifine lists, I extract popular keywords and add them to "word_list". 
# To ensure these are indeed crucial keywords, I check frequency of them.
word_list = []
Name = list(set(df_UNIACE["Name"].dropna())) 
for i_name in Name:
    words = underthesea.word_tokenize(i_name)
    for i_word in words:
        if len(i_word) >=3: # In order to uninclude meaningless words/characters such as "(", "-", "?",...)
            word_list.append(i_word)
dict_i = {i: word_list.count(i) for i in word_list}
dict_sort = {k: v for k, v in sorted(dict_i.items(), key=lambda item: item[1], reverse=True)}
dict_sort
```

```python
#Then, I generate column "Elements_in_Name" which consists of meaningful words from "Name" column
df_UNIACE["Elements_in_Name"] = df_UNIACE["Name"].apply(lambda x: underthesea.word_tokenize(x) if x is not None else x)
```

```python
# Define function for categorying
dict_category = {'Data analytics': ["dữ liệu", "chuẩn hóa", "Data", "phân tích", "Power BI", "DA", "Excel", "analytics",\
    "VBA", "DATA-DRIVEN", "Unpivot","pivot", "file", "where", "distinct", "Power query", "query", "hàm", "analysic"],
    'Management': ["Chief", "cấp cao", "cấp trên", "CEO", "MANAGEMENT"],
    'Life skill': ["life", "cuộc sống", "sống", "tư duy", "ý tưởng", "vượt qua", "áp lực", "bản thân", "cảm xúc", "ý nghĩa", "xây dựng",\
    "thời gian", "văn hóa"],
    'Sales': ["sales", "sale", "doanh số", "KPI", "bán hàng"],
    'Lập trình': ["bảo mật", "thông tin", "lập trình", "phần mềm", "ngôn ngữ", "programer", "developer", "công nghệ", "truy cập"],
    'Uniace Programe': ["uniace", "programe", "chương trình", "talent", "young"]
}
def category(words):
    if words is None: #Check for [None] in column "Name1" to correctly identify NaN values in Column "Name".
        return None
    else:
        for category, list in dict_category.items():
            if any(word.lower() in [i_list.lower() for i_list in list] for word in words):
                return category
                break
        return "Orthers"
```

```python
#Finally, Generate "Category" column which contains fields/topics users are interested in
df_UNIACE["Category"] = df_UNIACE["Elements_in_Name"].apply(category)
del df_UNIACE["Elements_in_Name"]
df_UNIACE["Category"].value_counts()
```

Generate column "SourcePlatform" identifying which browser user uses for accessing

```python
def get_browser_name(referrer_url):
    domain = urlparse(referrer_url).netloc
    if "google" in domain:
        return "Google"
    elif "bing" in domain:
        return "Bing"
    elif "coccoc" in domain:
        return "Coc Coc"
    elif "uniace" in domain:
        return "Uniace"
    elif "jotform" in domain:
        return "JotForm"
    elif "beacons" in domain:
        return "Beacons"
    elif "facebook" in domain:
        return "Facebook"
    elif "tiktok" in domain:
        return "Tiktok"
    else:
        return "Others"   
```

```python
df_UNIACE["SourcePlatform"] = df_UNIACE["MA_Referrer"].apply(lambda x: get_browser_name(x) if x is not None else "Uniace")
# MA_Referrer is None this means user goes directly to the website Uniace without going through a third party.
df_UNIACE["SourcePlatform"].value_counts()
```

Generate a new column called "Type_of_User" that identifies if user is worker or student based on email

```python
def Type_of_User(email):
    if re.search(r'\.edu\.',email):
        return "Student"
    elif re.search(r'@outlook', email):
        return "Worker"
    else:
        return "Orthers"
```

```python
df_UNIACE["Type_of_User"] = df_UNIACE["Email"].apply(lambda x: Type_of_User(x) if x is not None else None)
df_UNIACE["Type_of_User"].value_counts()
```

Also, I generate a new column called "School_of_student" that identifies which school user (if user is student) studies at based on email

```python
def get_school(email):
    if re.search(r'\.edu\.', email):
        email_elements = email.split("@")[-1].split(".")
        school = email_elements[email_elements.index("edu")-1]
        return school
    else:
        return None
```

```python
df_UNIACE["School_of_student"] = df_UNIACE["Email"].apply(lambda x: get_school(x) if x is not None else None)
df_UNIACE["School_of_student"].value_counts()
```

Generate a new column called "Province" that identifies which province user is lives in based on ip adress

```python
database_path = r'C:\Users\hungl\Desktop\PROJECT\UNIACE_project\GeoLite2-City.mmdb'
def get_province(ip):
    with geoip2.database.Reader(database_path) as reader:
        try:
            response = reader.city(ip)
            province = response.subdivisions.most_specific.name
            return province
        except:
            return "Invalid IP"
```

```python
df_UNIACE["Province"] = df_UNIACE["IP_Address"].apply(lambda x: get_province(x) if x is not None else None)
df_UNIACE["Province"].value_counts()
```
